# Reading the CSV File 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/AlvaDave/MTA-Inflation-Project/refs/heads/main/MTA_Price.csv"
df = pd.read_csv(url)
df.head()
```

# Cleaning Data

```python
# Renaming columns
df.columns = [
    'year',
    'base_fare',
    '30unlimited',
    '14unlimited',
    '7unlimited',
    '1unlimited']
df.head()
```

```python
# Remove Dollar Sign and Commas
df = df.replace(r"\$", "", regex=True)
df = df.replace(r",", "", regex=True)
df.head()
```

```python
# Remove Dollar Signs and extra commasfor Specific Columns
dollar_col  = ['base_fare',
    '30unlimited',
    '14unlimited',
    '7unlimited',
    '1unlimited']

for col in dollar_col: # creates loop for each column
    df[col] = (
        df[col] # overwrites the df[col] with df[col...]
            .astype(str) # converts to text to replace 
            .str.replace("$","")
            .str.replace(",","")
            .astype(float) # not int because integers can't have decimals. 
    )
df.head()    
```

```python
# Remove Extra Spaces 
df = df.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
```

```python
# Year Index
df['year'] = pd.to_numeric(df['year'], errors="coerce") # Clean Year Column, numeric groups 
df = df.set_index('year')
df = df.sort_index() # sort over time
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
import streamlit as st
import pandas as pd
import plotly.express as px



# title
st.title("MTA Fare Inflation Dashboard")

# year filter
year_range = st.slider(
    "Select Year Range",
    int(df.index.min()),
    int(df.index.max()),
    (2000, 2023)
)

filtered = df.loc[year_range[0]:year_range[1]]

# plot using index
fig = px.line(
    filtered,
    x=filtered.index,
    y="base_fare",
    title="Fare Over Time"
)

st.plotly_chart(fig)
```
